# Step 2: Import Data, Pase Date, and save it in a Dataframe then review outputs

```python
#Import Data, make sure the notebook is in the same directory as your notebook, otherwise, define a path to where the data is and use it to make the dataframe
mining_data=pd.read_csv('raw_data.csv', decimal=',',parse_dates=["date"],infer_datetime_format=True, sep=',')
mining_data.head()
```

```python
#check and remove negative values if existing
for cols in mining_data.columns.tolist()[1:]:
    df = mining_data.loc[mining_data[cols] > 0]
df.head()
df.info()
```

## Checkpoint #1: Based on the previous analysis, there are no missing values, data is fairly clean, but there are different scales for different columns and may require scaling

```python
df=df.set_index('date')
```

# Step 3: Visualize data distribution to identify whether outliers exist

```python
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use(['dark_background'])
df.hist(bins = 70, figsize = (17,15))
plt.show()
plt.suptitle('figure title', color='w')
```

```python
df.head()
```

# Step 4: Remove Outliers using z-score method, removing datapoints lying outside 2 standard deviations 

```python
df_copy=df
```

```python
#remove outliers outside 2 standard deviations from mean

from scipy import stats
df_copy=df[(np.abs(stats.zscore(df)) < 2).all(axis=1)]
df_copy.info()

#For each column, first it computes the Z-score of each value in the column, relative to the column mean and standard deviation.
#Then is takes the absolute of Z-score because the direction does not matter, only if it is below the threshold.
#all(axis=1) ensures that for each row, all column satisfy the constraint.
#Finally, result of this condition is used to index the dataframe.
```

```python
df_copy.head()
```

```python
plt.style.use(['dark_background'])
df_copy.hist(bins = 70, figsize = (17,15))
plt.show()
plt.suptitle('figure title', color='w')
df_copy.info()
```

# Step 5: Investigate and Decide on whether the data should be resampled or not

```python
#Seeing impact of resampling using only first sample of every hour on the distribution
df_first = df_copy.resample('H').first()
df_first.hist(bins = 70, figsize = (17,15))
plt.show()
plt.suptitle('figure title', color='w')
df_first.info()
```

```python
#Seeing impact of resampling using hourly mean on the distribution
df_mean = df_copy.resample('H').mean()
df_mean.hist(bins = 70, figsize = (17,15))
plt.show()
plt.suptitle('figure title', color='w')
df_mean.info()
```

## Checkpoint 2: A decision was made to keep all data minus outliers.  Resampling dramatically reduces amount of data for models to learn from (by factor of 180!). After removing outliers, there are roughly 346K data points left.

# Step 6: Features Extraction 

```python
#Dropping % Iron Concentrate Given that it is an output of laboratory analysis and highly cooliner to %Silica
df_copy.drop(columns=['% Iron Concentrate'],inplace=True)
df_copy.head()
```

```python
#Plot a heatmap to review correlations
import seaborn as sns
plt.figure(figsize=(20, 20))
p = sns.heatmap(ml_mining_data.corr(), annot=True);
```

```python
#MANUALLY EXTRACTING FEATURE

df=ml_mining_data.drop(['% Silica Feed', 'Starch Flow','Ore Pulp Flow','Flotation Column 01 Air Flow',
                        'Flotation Column 03 Air Flow', 'Flotation Column 04 Air Flow','Flotation Column 05 Air Flow',
                        'Flotation Column 01 Level','Flotation Column 02 Level', 'Flotation Column 03 Level', 
                        'Flotation Column 05 Level','Flotation Column 06 Level'], axis = 1) 


#Decision Criteria was based on
#1) correlation of feature to target variable (note some features with low correlation were included because of meeting criteria #3)
#2) removing cooliner data points to prevent overfitting
#3) Reviewing the distribution of features and ensuring the feature has contineous distribution. (note, some with bad distrib were kept because of #1)
        
#commented below are the columns to keep based on MANUAL Decisions on which columns to keep to and remove.
#ml_mining_data['% Iron Feed', 'Amina Flow','Ore Pulp pH', 'Flotation Column 02 Air Flow',
                          #   'Flotation Column 06 Air Flow','Flotation Column 07 Air Flow', 
                           #  'Flotation Column 04 Level', 'Flotation Column 07 Level',
                            # '% Silica Concentrate']
```

```python
# FEATURE ENGINEERING
#combining remaining air flow features by taking their mean, and similarly for the level data. This decision was made given how close the data was to one another
# This would also help later on when building the machine learning models given the less columns that have to be considered


df.head()

#new column names are representative of averages of column #s
df['avg_air_flow_267'] = df[['Flotation Column 02 Air Flow','Flotation Column 06 Air Flow','Flotation Column 07 Air Flow']].mean(axis=1)
df['avg_float_level_47'] = df[['Flotation Column 04 Level', 'Flotation Column 07 Level']].mean(axis=1)
df.head()
plt.figure(figsize=(20, 10))
p = sns.heatmap(df.corr(), annot=True);
```

```python
#Further cleaning of dataset and final view of correlations

df_ml=df.drop(['Flotation Column 02 Air Flow','Flotation Column 06 Air Flow','Flotation Column 07 Air Flow','Flotation Column 04 Level', 'Flotation Column 07 Level'], axis = 1) 
df_ml.head()
cols = df_ml.columns.tolist()
cols
cols = cols[-1:] + cols[:-1]
cols = cols[-1:] + cols[:-1]
df_ml = df_ml[cols]
df_ml.head()
df_ml.info()
plt.figure(figsize=(20, 8))
p = sns.heatmap(df_ml.corr(), annot=True);

df_ml.hist(bins = 70, figsize = (10,10))
plt.show()
plt.suptitle('figure title', color='w')
```
